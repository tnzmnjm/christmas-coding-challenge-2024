**Notes about Neural Networks**


*   Imagine our goal is to find the T-shirt with the highest probability of being a top seller T-shirt. In a NN with 2 layers, the inputs of the first layer are `price`, `shipping cost`, `marketing`, and `material`. There are 3 neurns in the first layer. The outputs of this layer are called **`activations`** and are `affordability`, `awareness`, and `perceived quality`.



*   The way neuron networks are implemented is that each neuron in each layer will have access to every feature from the previous layer.
*   We have the input layer, the hidden layer and the output layer. Its called the hidden layer because we see in the dataset what the `X` and `Y` are but do not see the values of affordability, awareness and perceived quality.
* The Neural networks **with multiple layers** are also called  `**Multilayer perceptrons**`. One thing we need to decide is the number of hidden layers and number of neurons in each layer.

* `Inference`: making predictions (`forward propagation`). It is called a forward propagation because you are calculating from the left to the right.
In contrast theres a `backward propagation` that is used for `learning`.
